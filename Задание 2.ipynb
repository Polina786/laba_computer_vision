Задание 17. Выполните стыковку тензоров a и b по вертикали.

Вход:

> a = torch.arange(2, 10, 2).reshape(2, -1)

> b = torch.repeat(1, 10).reshape(2, -1)

Желаемый результат: новый тензор размером $4 \times 5$.


```python
# ваш код здесь
a = torch.arange(2, 22, 2).reshape(2, -1)
b = torch.arange(1, 21, 2).reshape(2, -1)
t_17 = torch.cat([a, b], dim=0)
print(t_17)
```

Задание 18. Выполните стыковку тензоров a и b по горизонтали.

Вход:

a = torch.arange(2, 10, 2).reshape(2, -1)

b = torch.repeat(1, 10).reshape(2, -1)

Желаемый результат: новый тензор размером $2 \times 10$.

```python
# ваш код здесь
# Минимальное решение
a = torch.arange(2, 22, 2).reshape(2, -1)
b = torch.arange(1, 21, 2).reshape(2, -1)
t_18 = torch.cat([a, b], dim=1)
print(t_18)
```

Задание 19. Получите все элементы от 5 до 10 в тензоре a.

Вход:

> a = torch.tensor([2, 6, 1, 9, 10, 3, 27])

Желаемый результат:

> tensor([6, 9, 10])

```python
# ваш код здесь
a = torch.tensor([2, 6, 1, 9, 10, 3, 27])
result = a[(a >= 5) & (a <= 10)]
print(result)
```

Задание 23. Вставить в двумерный терзор из задания 17 строку из минимального элемента перед первой строкой.

```python
# ваш код здесь
min_element = t_17.min().item()
min_row = torch.full((1, 5), min_element)
t = torch.cat([min_row, t_17], dim=0)
print(t)
```

Задание 24.	В тензоре, полученном при выполнении задания 23, каждый элемент заменить его корнем квадратным.

```python
# ваш код здесь
t_24 = torch.sqrt(t)
print(t_24)
```

Задание 25.	Заменить в тензоре из задания 18 все значения, кратные 3 или 2 на число 1, а остальные – оставить без изменений.

```python
# ваш код здесь
mask = (t_18 % 2 == 0) | (t_18 % 3 == 0)
result = t_18.clone()
result[mask] = 1
print(result)
```
